```python
df1.drop(['Unnamed: 0','first','last','cc_num','street','trans_num'],inplace=True,axis=1)
```

```python
df1.info()
```

```python
df1.columns = df1.columns.str.lower().str.replace(" ", "_")
```

```python
df1['trans_date_trans_time']=pd.to_datetime(df1['trans_date_trans_time'])
```

```python
df1['hour']=df1['trans_date_trans_time'].dt.hour
df1['day']=df1['trans_date_trans_time'].dt.day
df1['month']=df1['trans_date_trans_time'].dt.month
```

```python
from datetime import datetime

df1['dob'] = pd.to_datetime(df1['dob'])
df1['age'] = (datetime.now() - df1['dob']).dt.days // 365
```

```python
df1['amount_category'] = pd.cut(df1['amt'],
                              bins=[0, 50, 200, 1000, 5000],
                              labels=['Low', 'Medium', 'High', 'Very High'])

df1['age_group'] = pd.cut(df1['age'],
                        bins=[18, 30, 45, 60, 100],
                        labels=['18-30', '30-45', '45-60', '60+'])
```

```python
df1['distance'] = np.sqrt((df1['lat'] - df1['merch_lat'])**2 + (df1['long'] - df1['merch_long'])**2)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.countplot(x='is_fraud', data=df1)
plt.title("Fraud vs Non-Fraud")
plt.show()
```

```python
sns.boxplot(x='is_fraud', y='amt', data=df1)
plt.title("Amount vs Fraud")
plt.show()
```

```python
plt.figure(figsize=(18,8))
sns.countplot(x='hour',hue='is_fraud',data=df1)
plt.title("Hour vs Fraud")
```

```python
sns.countplot(x='gender',hue='is_fraud',data=df1)
plt.title("Hour vs Fraud")
```

```python
top_cities=df1[df1['is_fraud']==1]['city'].value_counts().head(10)
top_cities.plot(kind='barh',figsize=(10,10))
```

```python
sns.countplot(y='category', hue='is_fraud', data=df1)
plt.title("Fraud by Category")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df1.corr(numeric_only=True), cmap='coolwarm')
```

```python
sns.countplot(x='amount_category',hue='is_fraud',data=df1)
```

```python
sns.boxplot(x='is_fraud', y='distance', data=df1)
plt.title("Distance vs Fraud")
plt.show()
```

```python
df1['fraud_label'] = df1['is_fraud'].map({0: 'Normal', 1: 'Fraud'})
```
